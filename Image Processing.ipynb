```python
#changing imgae format

from skimage.color import rgb2gray
img = imread('D:\\Users\\suraj_haradagatti\\Desktop\\PythonProjects\\MachineLearning\\Files\\images.jpg')
img_new = rgb2gray(img) # converts normal image to gray scale

plt.subplot(121), imshow(img)
plt.title('RGB Format') 

plt.subplot(122), imshow(img_new)
plt.title('Grayscale Format') 

plt.show()
```

```python
#adjust brightness of image 
# For gamma greater than 1, the output image will be darker than the input image. While for gamma less than 1, 
#the output image will be brighter than the input image.

from skimage import exposure

#adjusting brightness
image = imread('D:\\Users\\suraj_haradagatti\\Desktop\\PythonProjects\\MachineLearning\\Files\\images.jpg')
image_bright = exposure.adjust_gamma(image, gamma=0.5,gain=1)
image_dark = exposure.adjust_gamma(image, gamma=1.5,gain=1)

# plotting images
plt.subplot(131), imshow(image)
plt.title('Original Image')

plt.subplot(132),imshow(image_bright)
plt.title('Bright Image')

plt.subplot(133),imshow(image_dark)
plt.title('Dark Image')

plt.show()
```

```python
#using filters on images
# When we apply a filter on an image, every pixel value is replaced by a new value generated using surrounding pixel values. 
#The simplest filter is the median filter, where the pixel values are replaced with the median of neighboring pixels.

from skimage.filters import median

image = imread('D:\\Users\\suraj_haradagatti\\Desktop\\PythonProjects\\MachineLearning\\Files\\images.jpg', as_gray=True)
image_median = median(image)

# plotting images
plt.subplot(121), imshow(image)
plt.title('Original Image')

plt.subplot(122),imshow(image_median)
plt.title('Smooth Image')

plt.show()

```

```python
#sobel filter

from skimage.filters import sobel_h

image = imread('D:\\Users\\suraj_haradagatti\\Desktop\\PythonProjects\\MachineLearning\\Files\\images.jpg', as_gray=True)
image_sobelh = sobel_h(image)

# plotting images
plt.subplot(121), imshow(image)
plt.title('Original Image')

plt.subplot(122),imshow(image_sobelh, cmap = True)
plt.title('Horizontal Edge')

plt.show()
```
